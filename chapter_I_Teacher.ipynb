### Continuous vs discrete signals
A continuous signal is defined in all instant of time:
$$
x(t),t \in \mathbb{R}
$$

Meanwhile a discrete signal is defined in specific instants of time:
$$
x[n], n \in A, A\subset \mathbb{Z}
$$

## What is the behaviour of frequency in continuous domain?


The essential cases for analysis of periodic signals are sinusoidal signals and harmonically related complex exponential signals.
Continuous sinusoidal signal over time:
$$
x(t)=A \cos (\Omega t+\Theta), \Omega=2 \pi F
$$

Properties
* Periodicity $x\left(t+\frac{1}{F}\right)=x\left(t\right)$:

* Multiple sinusoidal signals with different frequencies are different from each other.
These properties are also applied to complex exponential signals.

### Euler identity
$$
Ae^{j(\Omega t + \Theta)} = A\cos(\Omega t + \Theta) + jA\sin(\Omega t + \Theta), \\
A\cos(\Omega t + \Theta) = \frac{A}{2} e^{j(\Omega t + \Theta)} + \frac{A}{2}e^{-j(\Omega t + \Theta)}, \\ A\sin(\Omega t + \Theta) = \frac{A}{2j} e^{j(\Omega t + \Theta)} - \frac{A}{2j}e^{-j(\Omega t + \Theta)}
$$
For mathematical convention, the frequency of a signal can be positive or negative.


```python
#Continuous signal over time
t = arange(0,10,0.01)
w1=np.pi/2 ; w2=np.pi ; w3=np.pi*3/2 ;
y1 = np.sin(w1*t); y2 =np.sin(w2*t) ; y3= np.sin(w3*t)
plt.figure(figsize=(10,7));
plt.suptitle("Frequency in continuous domain");
plt.subplot(2,2,1);plt.plot(t, y1);plt.grid(True);plt.xlabel("Time (s)");plt.ylabel("Amplitude")
plt.subplot(2,2,2);plt.plot(t, y2);plt.grid(True);plt.xlabel("Time (s)");plt.ylabel("Amplitude")
plt.subplot(2,2,3);plt.plot(t, y3);plt.grid(True);plt.xlabel("Time (s)");plt.ylabel("Amplitude")
plt.subplot(2,2,4);plt.plot(t, y1);plt.plot(t, y2);plt.plot(t, y3);plt.grid(True);plt.xlabel("Time (s)");plt.ylabel("Amplitude")
plt.legend(("w1 = pi/2", "w2 = pi", "w3 = 3pi/2"));
```

## How is the behaviour of the frequency in harmonically related complex signals?

Harmonically related complex exponential signals continuous over time:
$$
s_k(t) = e^{jk\Omega_ot},\quad \Omega_o = 2\pi F_o,\space F_o\text{: Fundamental frequency, }\space k\in \mathbb{Z}
$$

##### **Properties**
* Each  $s_k \left(t\right)$ is characterized by a ${\mathrm{kF}}_o$  (harmonically related set of signals $s_k$ to fundamental frequency $f_o$).
* Considering that each $s_k \left(t\right)$  has a period value of $\frac{1}{{\mathrm{kF}}_o }$ , all the set of signals are periodic for $T=\frac{\;1}{F_o }$
* $s_{\mathrm{k1}} \left(t\right)\not= s_{\mathrm{k2}} \left(t\right)$ , if $\mathrm{k1}\not= \mathrm{k2}$ due to having different frequencies

Find the fundamental period of the signal $s_k \left(t\right)=e^{\mathrm{jk}\frac{\;5\pi \;}{2}t} ,k\in \mathbb{Z}$.

```python

t = np.arange(0, 2, 0.01)
s1 = np.real(np.exp(1j*5*np.pi/2*1*t))
s2 = np.real(np.exp(1j*5*np.pi/2*2*t))
s3 = np.real(np.exp(1j*5*np.pi/2*3*t))

plt.figure(16)
plt.plot(t, s1, label="k = 1")
plt.plot(t, s2, label="k = 2")
plt.plot(t, s3, label="k = 3")
plt.xlabel("t")
plt.grid(True)
plt.legend()
```

The fundamental period is $0.8$.

Harmonically related complex exponential signals discrete over time:
$$s_k[n] = e^{jk\omega_o n},\quad \omega_o = 2\pi f_o,\space f_o\text{: Fundamental frequency,}\space k \in \mathbb{z}$$

##### Properties
* Assuming an integer period of N samples, $f_o =\frac{\;1}{N}$:
$$s_k[n] = e^{j\frac{2\pi k}{N}n}$$

* Then, there are only  harmonically related complex exponential signals. The rest are replicas (**Alias**).
$$s_{k+N}[n] = e^{j\frac{2\pi (k+N)}{N}n} = e^{j\frac{2\pi k}{N}n} = s_k[n]$$

The discrete signal $s_k \left\lbrack n\right\rbrack =e^{j\frac{\;2\pi \;k}{N}n}$, has a period of $N=3$. Considering $k=1,4\;\mathrm{and}\;7$ , find the repetitions of the signal.

```python
N = 3
n = np.arange(0, 11)
s1 = np.real(np.exp(1j*2*np.pi*1/N * n))
s4 = np.real(np.exp(1j*2*np.pi*4/N * n))
s7 = np.real(np.exp(1j*2*np.pi*7/N * n))

plt.figure(17)
plt.stem(n, s1, linefmt='C0-', markerfmt='C0o', basefmt='C0-', label="s1")
plt.stem(n, s4, linefmt='C1-', markerfmt='C1o', basefmt='C1-', label="s4")
plt.stem(n, s7, linefmt='C2--', markerfmt='C2o', basefmt='C2--', label="s7")
plt.grid(True)
plt.legend()    
```

## Sampling


**Periodic sampling**:
$$x[n] = x(nT),\quad n\in \mathbb{Z} \\ F_s = \frac{1}{T_s},\quad \text{Sampling frequency}$$
For periodic signals,  $F_s$  give a direct relation between $\Omega$ from the continuous version of $x\left(t\right)$ and  from the discrete version $x\left\lbrack n\right\rbrack$ .
$$\omega = \frac{\Omega}{F_s} \rightarrow f = \frac{F}{F_s}$$

Considering that $-\pi \le \omega \le \pi$ is the fundamental range, $F$ must comply one condition for not be an Alias:
$$-\frac{F_s}{2} \leq F \leq \frac{F_s}{2}$$
Find the normalized frequency of $x\left(t\right)$:
$$x(t) = 5\cos(2\pi (10) t + \frac{\pi}{4}); \quad T_s = 0.025$$

```python
Ts = 0.025
t = np.arange(0, 0.4 + Ts, Ts)
Fs = 1/Ts
F = 10
x = 5*np.cos(2*np.pi*F*t + np.pi/4)
n = np.arange(0, len(t))
xn = 5*np.cos(2*np.pi*F/Fs*n + np.pi/4)
FN = F/Fs
plt.figure()
plt.stem(t, x)
plt.xlabel("Time(s)")
plt.title("Discrete-time signal")
plt.grid(True)
```

If the previous condition is not met, the resultant discrete signal is affected by the Aliasing efect. Then, a composed signal by one or more periodic signals and a sampling frequency ($F_s$) is possible obtain its discrete version free from ambiguity if its higher frequency ($F_{\max }$) follow the inequality:
$$F_{max} \leq \frac{F_s}{2}$$



### Sampling Theorem (Nyquist Theorem)
* Considering an analog signal with a high frequency value of  $F_{\max }$ is possible obtain a digital signal without ambiguity generated by the inclusion of Alias (**Aliasing**).
* This condition will be  $2F_{\max } =F_N \le F_s$ where $F_N$ is the Nyquist frequency.

## What is the classification of discrete-time systems?

* **Static (memory-less)**: it does not depend on past nor future input samples.
* **Dynamic (with memory)**: depend on pass or future input samples
* **Time invariant**: an input delay generates an output delay
* **Linear**: superposition principle  addition and scalability
* **Causality**: the output only depends on the current and previous inputs
* **Bounded Input, Bounded Output (BIBO) stable**: if the input is bounded
